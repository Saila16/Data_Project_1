## **1. Merging the data into one dataframe**
a. First, merge orders_data and order_item_data
b. Finally, Merge the first step with product_data and customer_data

```python
#Merger order data dengan order item data
mergedata1 = order_item_data.merge(orders_data, how='left',on='order_id')
print('Dimension', mergedata1.shape)
mergedata1.head()
```

```python
#merger semua data
mergedata2 = mergedata1.merge(products_data, how='left', on='product_id')
print("Dimension" , mergedata2.shape) 
mergedata2.head()
```

```python
#merger semua data (jumlah baris tetap 112650, kolom bertambah)
final_data= mergedata2.merge(customers_data, how='left' , on='customer_id')
print("Dimension" , final_data.shape)
final_data.head()
```

## **4. In the merged dataframe, please add new column to get the date and the time (hour) and day name from:**
- order_purchase_timestamp
    - dt.hour -> jam
    - dt.day_name -> hari
- order_delivered_customer_date


```python
final_data.head()
```

```python
final_data['order_purchase_timestamp'].dt.hour
```

```python
final_data['order_purchase_timestamp'].dt.day_name()
```

```python
#mencari jam dan hari
final_data['order_purchase_hour'] = final_data['order_purchase_timestamp'].dt.hour
final_data['order_purchase_day_name'] = final_data['order_purchase_timestamp'].dt.day_name()
final_data.head()
```

```python
#order_delivered_customer_date	
final_data['order_delivered_hour'] = final_data['order_delivered_customer_date'].dt.hour
final_data['order_delivered_day_name'] = final_data['order_delivered_customer_date'].dt.day_name()
final_data.head()
```

## **5.How many unique customer do the transaction in each city?**
- Clue, use pivot table or group by

```python
hasil_pivot = pd.pivot_table(final_data, values='customer_id' , index='customer_city', aggfunc=pd.Series.nunique)
hasil_pivot
```

## **6. Get the top 5 cities that has the highest number of unique customer**
- clue, sorting the data using sort_values

```python
#data 5 terbesar
Get_Top_5= hasil_pivot.reset_index().sort_values(by='customer_id', ascending = False).reset_index(drop=True)
Get_Top_5[:5]
```

## **7. At the 5 cities, what time the customer do the transaction? show them in :**
- Average
- Median
- Clue, use pivot table or  group by

```python
g = final_data.groupby('customer_city').agg(customer_id=('customer_id' , 'count'), average_order=('order_purchase_hour', 'mean'), median_order=('order_purchase_hour', 'median'))
g.reset_index().sort_values(by=['customer_id'], ascending=False).head(5).reset_index(drop=True)
```
